```python
import pandas as pd
import os
```

```python
def name_company_split(name):
    undef_list = ['Unidentified Company Representative',
                  'Unidentified Corporate Participant',
                  'Unidentified Analyst']
    undef_in = False
    for undef_id in undef_list:
        if undef_id in name:
            result_list = [undef_id, 'Unidentified Company']
            undef_in = True
            break
    if not undef_in:
        utf_name = name.encode('utf-8')
        result_list = [name, 'No_company_found']
        if b'\xe2\x80\x93' in utf_name:
            result_list = utf_name.split(b'\xe2\x80\x93')
            result_list = [result_list[0].decode(), result_list[1].decode()]
        elif '-' in name:
            result_list = name.split('-')
        result_list = [result_list[0], ' - '.join(result_list[1:])]
    return list(map(lambda x: x.strip(), result_list))
```

```python
def _row_proc(row):
    analytic = row['Analyst']
    an_list = row['Analysts_list']
    head_analytics_str = an_list.replace("'", "").split(", ")
    head_analytics_list = list(map(name_company_split, head_analytics_str))
    for i in range(len(head_analytics_list)):
        if analytic.lower().replace(" ", "")==head_analytics_list[i][0].lower().replace(" ", ""):
            return head_analytics_list[i][1]
    return "Nothing_found"
```

```python
path = "../data/res_tables/"
```

```python
files = os.listdir(path)
```

```python
######
```

```python
res_analysts = []
res_banks = []
for j in range(len(files)):
    print(j)
    df = pd.read_csv(path+files[j]).dropna()
    for f in pd.unique(df['File_path']):
        cur_df = df[df['File_path']==f].copy()
        analytics_and_banks = cur_df[['Analyst', 'Analysts_list']].drop_duplicates()
        unique_banks = pd.unique(analytics_and_banks.apply(_row_proc, axis=1)).tolist()
        unique_analysts = pd.unique(analytics_and_banks['Analyst']).tolist()
        if "" in unique_banks:
            unique_analysts.remove("")
        if "" in unique_banks:
            unique_analysts.remove("")

        res_banks += unique_banks
        res_analysts += unique_analysts
```

```python
pd.DataFrame(pd.unique(res_banks)).to_excel("banks.xlsx")
```

```python
pd.DataFrame(pd.unique(res_analysts)).to_excel("analysts.xlsx")
```

```python
path = "../data/result_tables/names_tables/"
files = os.listdir(path)
```

```python
res_analysts = []
res_banks = []
for j in range(len(files)):
    print(j)
    df = pd.read_excel(path+files[j]).dropna()
    for f in pd.unique(df['File_path']):
        cur_df = df[df['File_path']==f].copy()
        ab_str = cur_df['Analysts_list'].values[0].replace("{", "").replace("}", "")
        if (len(ab_str)<650) & (len(ab_str)>0):
            to_list = ab_str.split("', '")
            a_bll = list(map(lambda x: x.replace("'", "").split(": "), to_list))
            
            append_list_bank = list(map(lambda x: x[1], a_bll))
            if "" in append_list_bank:
                append_list_bank.remove("")
            res_banks += append_list_bank
            
            append_list_analytic = list(map(lambda x: x[0], a_bll))
            if "" in append_list_analytic:
                append_list_analytic.remove("")
            res_analysts += append_list_analytic
```
